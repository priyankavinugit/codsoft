# Data Cleaning

```python
#Delete unnecessary columns
data.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
data.head()
```

# Encode the data

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data['Gender']=labelencoder.fit_transform(data['Gender'])
data['Geography']=labelencoder.fit_transform(data['Geography'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
x=data.drop(columns='Exited')
y=data['Exited']
```

```python
x
```

```python
y
```

# Split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=51)
```

# Building Algorithms

## Random Forest CLassification

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

# Predict the Random Forest model

```python
y_pred=model.predict(x_test)
y_pred
```

```python
len(y_pred)
```

# Accuracy of Random Forest model

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,auc
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:{:.2f}%".format(accuracy*100))
```

# Visualization of RF Model using roc_curve

```python
y_prob=model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_prob)
plt.figure(figsize=(10,6))
plt.plot(fpr,tpr,color='red',lw=2)
plt.plot([0,1],[0,1],color='blue',lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title('RandomForestClassifier\nAccuracy:{:.2f}%'.format(accuracy*100))
plt.show()          
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

# Check the score of Logistic Regression model

```python
model_score=model.score(x_train,y_train)
print(model_score)
```

# Predict the Logistic Regression Model

```python
y_pred=model.predict(x_test)
print(y_pred)
```

# Accuracy of Logistic Regression Model

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:{:.2f}%".format(accuracy*100))
```

# Visualize the Logistic Regression Model using roc_auc curve

```python
y_prob=model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_prob)
plt.figure(figsize=(10,6))
plt.plot(fpr,tpr,color='red',lw=2)
plt.plot([0,1],[0,1],color='blue',lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title('Logistic Regression\nAccuracy:{:.2f}%'.format(accuracy*100))
plt.show()          
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier()
model.fit(x_train,y_train)
```

# Check the score of Gradient Boosting Classification Model

```python
model_score=model.score(x_train,y_train)
print(model_score)
```

# Predict the score of Gradient Boosting Classification model

```python
y_pred=model.predict(x_test)
print(y_pred)
```

# Accuracy of Gradient Boosting Classification Model

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:{:.2f}%".format(accuracy*100))
```

# Visualize the Gradient Boosting Classification Model using roc_auc Curve

```python
y_prob=model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_prob)
plt.figure(figsize=(10,6))
plt.plot(fpr,tpr,color='red',lw=2)
plt.plot([0,1],[0,1],color='blue',lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title('Gradient Boosting Classification\nAccuracy:{:.2f}%'.format(accuracy*100))
plt.show()          
```

# Comparing all the models

```python
model_names=['Random Forest','Logistic Regression','Gradient Boosting']
acc_list=[]
for model in [RandomForestClassifier(),LogisticRegression(),GradientBoostingClassifier()]:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    acc_list.append(accuracy)
```

```python
model_results=pd.DataFrame({"Model":model_names,"Accuracy":acc_list})
model_results
```

```python
sns.barplot(x="Model",y="Accuracy",data=model_results)
```
